```python
# 손실함수(Loss function)는 머신러닝이나 딥러닝 모델이 얼마나 잘 예측하고 있는지를 수치적으로 표현해주는 함수
#
# Robust fitting은 이상치(outlier)가 있는 데이터셋에서도 신뢰성 있게 모델을 적합시키는 방법
#
# scipy.optimize.least_squares 에서 loss function를 설정하면 Rubust fitting를 할수 있다
# 1. 'linear': 표준제곱합--> 일반 least squares
# 2. 'soft_l1': Huber 손실의 연속 완화 버전, 이상치에 좀 더 강건
# 3. 'huber':  Huber loss, MSE와 MAE의 중간 성질
# 4. 'cauchy': 이상치에 매우 강한 완화된 로지스틱 손실
# 5. 'arctan': 큰 오차를 더욱 억제하는 손실 함수
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import least_squares

# 1. 데이터 생성
np.random.seed(42)
x = np.linspace(0, 2, 5000)
A_true, tau_true, C_true = 50, 0.5, 1
y_true = A_true * np.exp(-x / tau_true) + C_true

# 2. 노이즈와 outlier 추가
noise = np.random.normal(0, 10, size=len(x))  # 큰 노이즈
y_noisy = y_true + noise

out_size = 1000

outlier_idx = np.random.choice(len(x), size=out_size, replace=False)
y_noisy[outlier_idx] += np.random.normal(100, 5, size=out_size)

# 3. 모델 정의
def model(x, A, tau, C):
    return A * np.exp(-x / tau) + C

def residuals(params, x, y):
    A, tau, C = params
    return model(x, A, tau, C) - y

# 4. robust fitting using least_squares
init_guess = [10, 2, 0]
res1 = least_squares(residuals, init_guess, args=(x, y_noisy), loss='linear')
res2 = least_squares(residuals, init_guess, args=(x, y_noisy), loss='soft_l1')
res3 = least_squares(residuals, init_guess, args=(x, y_noisy), loss='huber')
res4 = least_squares(residuals, init_guess, args=(x, y_noisy), loss='cauchy')
res5 = least_squares(residuals, init_guess, args=(x, y_noisy), loss='arctan')

A_fit1, tau_fit1, C_fit1 = res1.x
A_fit2, tau_fit2, C_fit2 = res2.x
A_fit3, tau_fit3, C_fit3 = res3.x
A_fit4, tau_fit4, C_fit4 = res4.x
A_fit5, tau_fit5, C_fit5 = res5.x
# print(f"Estimated A={A_fit:.3f}, tau={tau_fit:.3f}, C={C_fit:.3f}")

# 5. 시각화
plt.figure(figsize=(10, 6))
plt.scatter(x, y_noisy, color='gray', label='data', alpha=0.5)
plt.scatter(x[outlier_idx], y_noisy[outlier_idx], color='tab:orange', label='outliers', alpha=0.5)
plt.plot(x, y_true, 'k--', label='True model', lw=2)
plt.plot(x, model(x, *res1.x), '--', label='Robust Fit (linear)', linewidth=2)
plt.plot(x, model(x, *res2.x), '--', label='Robust Fit (soft_l1)', linewidth=2)
plt.plot(x, model(x, *res3.x), '--', label='Robust Fit (huber)', linewidth=2)
plt.plot(x, model(x, *res4.x), '--', label='Robust Fit (cauchy)', linewidth=2)
plt.plot(x, model(x, *res5.x), '--', label='Robust Fit (arctan)', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential Decay Fit with Heavy Noise and Outliers')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
